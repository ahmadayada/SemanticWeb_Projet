```python
def extract_date_from(url) :
    url.split("=")
    jour = url.split("=")[2][0:2]
    mois= url.split("=")[3][0:2]
    annee = url.split("=")[-1]
    day = jour +'-'+ mois +'-'+ annee #to have the same format as the previous csv #"15-11-2021"
    return day

extract_date_from('https://www.meteociel.fr/temps-reel/obs_villes.php?code2=7475&jour2=16&mois2=10&annee2=2021')
```

```python

from rdflib import Graph

g = Graph()

schema = Namespace('http://schema.org/')
sosa = Namespace('http://www.w3.org/ns/sosa/')
for index, row in df1.iterrows() :
    jour = f'Day_{row["day"]}_{row["time"][0]}'
    #day = f'Day_{row["day"]}' #.rsplit("/",1)[-1]}
    g.add((URIRef(schema+f'{jour}/'), URIRef(schema+'time'), Literal(row['time'][0], datatype=XSD.dateTime)))
    g.add((URIRef(schema+f'{jour}/'), URIRef(schema+'outsideTemp'), Literal(row["outsideTemp"][0], datatype=XSD.float)))
    #g.add((URIRef(schema+f'{jour}/'), URIRef(schema+'outsideHmdt'), Literal(row['humidity'], datatype=XSD.string)))
    g.serialize('meteoNov.ttl',format='turtle')
```
