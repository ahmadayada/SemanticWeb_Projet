```python
import pandas as pd
import re
from rdflib import Graph, Literal, RDF, URIRef, Namespace #basic RDF handling
from rdflib.namespace import FOAF , XSD, SOSA
```

```python
url15 = 'https://www.meteociel.fr/temps-reel/obs_villes.php?code2=7475&jour2=15&mois2=10&annee2=2021'
dfs15 = pd.read_html(url15)
df15 = dfs15[4]

masque=df15[0].str.contains('^\d+ h', na=False)
df15=df15[masque]
df15=df15[[0,4]]
df15.columns=["time","outsideTemp"]
```

```python
df15['outsideTemp'][7]='7.0 °C'
df15
```

```python
heures = []
outsideTemps = []
for index, row in df15.iterrows() :
    #time = row['time']
    
    heure = re.findall("(\d+) h",row['time'])
    outsideTemp = re.findall("(\d.?\d) °C",row['outsideTemp'])
    #heure = heure.split("[")
    heures.append(heure[0])
    outsideTemps.append(outsideTemp[0])
    #index+=1
#print(heures)
#print(outsideTemps)
df15['time']=heures
df15['outsideTemp']=outsideTemps
df15

```

```python
#df['day'] = [extract_date_from(url)]*len(df.index)
df15['day'] = ['2021-11-15']*len(df15.index)
df15
df15.to_csv("meteo15Nov.csv",header= df15.columns , index=False)
```

```python
#"2021-11-15T06:32:52"^^xsd:dateTime
from rdflib import Graph

g = Graph()

schema = Namespace('http://schema.org/')
sosa = Namespace('http://www.w3.org/ns/sosa/')

for index, row in df15_.iterrows() :
    jour = f'Day_{row["day"]}_{row["time"]}'
    #day = f'Day_{row["day"]}' #.rsplit("/",1)[-1]}
    g.add((URIRef(schema+f'{jour}/'), URIRef(schema+'time'), Literal(row['dateTime'], datatype=XSD.dateTime)))
    g.add((URIRef(schema+f'{jour}/'), URIRef(schema+'outsideTemp'), Literal(row["outsideTemp"], datatype=XSD.float)))
    #g.add((URIRef(schema+f'{jour}/'), URIRef(schema+'outsideHmdt'), Literal(row['humidity'], datatype=XSD.string)))
    g.serialize('meteo15Nov_.ttl',format='turtle')
```
